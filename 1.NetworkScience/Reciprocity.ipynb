```python
import networkx as nx
import matplotlib.pyplot as plt
```

```python

def draw_network_colored(g, p, colors = 'default', fs = (8,4)):

    if colors == 'default': colors = {i:0 for i in g.nodes()}
    plt.figure(figsize = fs)
    nx.draw_networkx(g,
                     pos=p,
                     font_size = 12,
                     font_weight = 'bold',
                     font_color = 'plum',
                     edge_color = 'grey',
                     arrowsize = 20
                    )

    nodes = nx.draw_networkx_nodes(g, p,cmap=plt.get_cmap('Blues') ,node_color= [c for c in colors.values()],node_size = 300)
    nodes.set_edgecolor('lightgrey')
    plt.axis('off')
    plt.show()
```

```python
exercise_graph = nx.DiGraph()
exercise_graph.add_nodes_from([1,2,3,4,5,6])
exercise_graph.add_edges_from([(1,2),(3,1),(1,5),(2,1),(2,3),(4,3),(3,4),(7,5),(3,5),(5,6),(5,7)])
```

```python
layout = nx.circular_layout(exercise_graph)
draw_network_colored(exercise_graph,layout)
```
